```python
# get the data

data1 = np.loadtxt("withoutgpu_basic/totalenergy.dat",usecols = range(3))
#data2 = np.loadtxt("earth_sun_16/totalenergy.dat",usecols = range(3))
#data3 = np.loadtxt("earth_sun_8/totalenergy.dat",usecols = range(3))

kinetic1 = data1.T[0]
potential1 = data1.T[1]
total1 = data1.T[2]

#kinetic2 = data2.T[0]
#potential2 = data2.T[1]
#total2 = data2.T[2]

#kinetic3 = data3.T[0]
#potential3 = data3.T[1]
#total3 = data3.T[2]

l1 = len(total1)
#l2 = len(total2)
#l3 = len(total3)

timeunit1 = 1*0.08*0.0166
time1 = np.linspace(1,l1-1,l1)*timeunit1

#timeunit2 = 1*0.06*0.179
#time2 = np.linspace(1,l2-1,l2)*timeunit2

#timeunit3 = 1*0.12*0.179
#time3 = np.linspace(1,l3-1,l3)*timeunit3

print(timeunit1) #,timeunit2,timeunit3)
print(l1)
```

```python
# plot the results

plt.figure(figsize=(8,6))
plt.plot(time1, kinetic1, 'r', label = 'KE 0.005')
plt.plot(time1, potential1, 'g', label = 'PE 0.005')
plt.plot(time1, total1, 'b', label = 'TE 0.005')
#plt.plot(time2, kinetic2, 'g', label = 'KE 0.01')
#plt.plot(time2, potential2, 'g', label = 'PE 0.01')
#plt.plot(time2, total2, 'g', label = 'TE 0.01')
#plt.plot(time3, kinetic3, 'b', label = 'KE 0.02')
#plt.plot(time3, potential3, 'b', label = 'PE 0.02')
#plt.plot(time3, total3, 'b', label = 'TE 0.02')
plt.xlabel('time(Myr)')
plt.ylabel('energy(nbody unit)')
plt.title('Energy of star cluster')
plt.legend()
plt.show()
```

```python
# normalize the results

normkin1 = kinetic1/kinetic1[0]
normpot1 = potential1/potential1[0]
normtot1 = total1/total1[0]

#normkin2 = kinetic2/kinetic2[0]
#normpot2 = potential2/potential2[0]
#normtot2 = total2/total2[0]

#normkin3 = kinetic3/kinetic3[0]
#normpot3 = potential3/potential3[0]
#normtot3 = total3/total3[0]
```
